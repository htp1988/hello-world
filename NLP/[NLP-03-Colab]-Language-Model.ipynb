```python
import string

PUNC = list(string.punctuation)
```

```python
def preprocess(sent):
  for i in PUNC:
    sent = sent.replace(i, " ")
  return sent.lower()
```

```python
import nltk
nltk.download('reuters')
nltk.download('punkt')
from nltk.corpus import reuters
from nltk import bigrams, trigrams, pad_sequence
from collections import Counter, defaultdict
```

```python
sentences = []
[sentences.append(preprocess(sent).split()) for sent in corpus]
sentences
```

```python
model = defaultdict(lambda: defaultdict(lambda: 0))

for sentence in sentences:
    for w1, w2 in bigrams(sentence, pad_right=True, pad_left=True, left_pad_symbol="<s>", right_pad_symbol="</s>"):
        model[(w1)][w2] += 1
```

```python
model
```

```python
for w1 in model:
    total_count = float(sum(model[w1].values()))
    for w2 in model[w1]:
        model[w1][w2] = (model[w1][w2]+1)/(total_count+len(model))
```

```python
model
```

```python
test = ['<s>','trăm', 'năm', 'trong', 'cõi', 'người', 'ta', '</s>']
```

```python
pro = 1
for idx in range(len(test)-1):
  pro *= model[test[idx]][test[idx+1]]
pro
```

```python
import random

text = ["kiều"]
sentence_finished = False

while not sentence_finished:
  # select a random probability threshold
  r = random.random()
  accumulator = .0

  for word in model[text[-1]].keys():
      accumulator += model[text[-1]][word]
      # select words that are above the probability threshold
      if accumulator >= r:
          text.append(word)
          break

  if text[-1] == "</s>":
      sentence_finished = True

print (' '.join([t for t in text if t]))
```

#2.Neural Probabilistic Language Model
Implementing Bengio’s Neural Probabilistic Language Model (NPLM)

[Paper]https://www.jmlr.org/papers/volume3/bengio03a/bengio03a.pdf

```python
import string as pystring
from nltk import everygrams
```

```python
PUNCT_TO_REMOVE = pystring.punctuation + pystring.digits + "\n"
def clean_text(text):
    """custom function to removal: punctuations and digits"""
    text = text.translate(str.maketrans(' ', ' ', PUNCT_TO_REMOVE))
    text = text.lower()
    return text
clean_text(corpus[0])
```

```python
def get_contexts_and_labels(corpus):
    contexts, labels = [], []
    for line in corpus:
        line = clean_text(line)
        line = line.split()
        if len(line) < 2:
            continue
        ngrams = list(everygrams(line, min_len=2, max_len=len(line)))
        for gram in ngrams:
            context = " ".join(list(gram[:-1]))
            label = gram[-1]
            contexts.append(context)
            labels.append(label)
    return contexts, labels
```

```python
contexts, labels = get_contexts_and_labels(corpus)
```

```python
from sklearn.model_selection import train_test_split

train_sentences, valid_sentences, train_labels, valid_labels = train_test_split(contexts, labels, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
tokenizer = Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts(train_sentences)
```

```python
vocab_size = len(tokenizer.index_word) + 1
embedding_dim = 200
max_length = max([len(context.split()) for context in contexts])
```

```python
train_seq = tokenizer.texts_to_sequences(train_sentences)
train_seq_pad = pad_sequences(train_seq, maxlen=max_length, truncating='post', padding="post")

valid_seq = tokenizer.texts_to_sequences(valid_sentences)
valid_seq_pad = pad_sequences(valid_seq, maxlen=max_length, truncating='post', padding="post")
```

```python
def convert_labels(labels):
    convert_labels = []
    for label in labels:
        if label in tokenizer.index_word.values():
            convert_labels.append(to_categorical( tokenizer.word_index[label], len(tokenizer.word_index)+1))
        else:
            convert_labels.append(to_categorical( tokenizer.word_index["<OOV>"], len(tokenizer.word_index)+1))
    return convert_labels
```

```python
train_labels = convert_labels(train_labels)
valid_labels = convert_labels(valid_labels)
```

```python
train_labels = np.array(train_labels)
valid_labels = np.array(valid_labels)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense
```

```python
model_nplm = Sequential()
model_nplm.add(Embedding(vocab_size, embedding_dim, input_length=max_length))
model_nplm.add(Flatten())
model_nplm.add(Dense(10, activation='relu'))
model_nplm.add(Dense(vocab_size, activation='softmax'))
model_nplm.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_nplm.summary()
```

```python
model_nplm.fit(train_seq_pad, train_labels, epochs=30, validation_data=(valid_seq_pad, valid_labels))
```

```python
def generate_text(sample_text, number_words):
    list_words = sample_text.split()
    for i in range(number_words):
        sample_seq = tokenizer.texts_to_sequences([" ".join(list_words)])
        sample_seq_pad = pad_sequences(sample_seq, maxlen=max_length, truncating='pre', padding="post")
        predict_word = tokenizer.index_word[np.argmax(model_nplm.predict(sample_seq_pad))]
        list_words.append(predict_word)
    return " ".join(list_words)
```
